```python
#label encoding of string values
from sklearn.preprocessing import LabelEncoder
```

```python
eOutlook = LabelEncoder()
eTemp = LabelEncoder()
eHumidity = LabelEncoder()
eWind = LabelEncoder()
```

```python
data['Outlook_int']  = eOutlook.fit_transform(data['Outlook'])
data['Temp_int']  = eTemp.fit_transform(data['Temp'])
data['Humidity_int']  = eHumidity.fit_transform(data['Humidity'])
data['Wind_int']  = eWind.fit_transform(data['Wind'])
```
