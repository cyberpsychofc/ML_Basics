```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
# Lemmetization
for i in range(len(corpus)):
  r = re.sub('[^a-zA-Z]',' ', corpus[i][0])
  r = r.lower()
  r = r.split()
  r = [lem.lemmatize(word) for word in r if not word in set(stopwords.words('english'))]
  r = ' '.join(r)
  corpus.append(r)
```

```python
#Spam-classification
import pandas as pd
messages = pd.read_csv('SMSSpamCollection',sep='\t',names=["label","message"])
```

```python
messages.head()
```

```python
messages.shape
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z0-9]',' ', messages['message'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True,ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
#label encoding from ham/spam to 0/1
y = pd.get_dummies(messages['label'])
y = y.iloc[:,1].values
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, ytest = train_test_split(X,y,test_size=0.20,random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(ytest, y_pred)
```

```python
score
```

```python
print(classification_report(y_pred, ytest))
```

```python
#random-forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(accuracy_score(y_pred,ytest))
print(classification_report(y_pred, ytest))
```

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z0-9]',' ', messages['message'][i])
  review = review.lower()
  review = review.split()

  review = [lem.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
#training model from scratch
model = gensim.models.Word2Vec(words,window=5,min_count=2)
```

```python
import numpy as np
def avgw2v(doc):
  #remove OOV words
  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```
